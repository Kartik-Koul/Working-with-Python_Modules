```python
# First of all, we are talking about matplot.collections.PatchCollection.
# Before heading straight to the code, just a brief on what Patches are.
# A patch is basically a 2D shape. Common examples are a rectangle, a circle, polygons.
```

```python
# PatchCollection is an easy way to group multiple patch objects together and give them similar properties.
# To simply put, instead of individually initializing a large number of patch objects separately, we can do it at once.
# Makes our work easier, doesn't it? Fair warning - These collections are not as flexible as individual patch objects.
# However, they are faster to work with for common use cases and are thus often used.
```

```python
import matplotlib.pyplot as plt # Good old pyplot
from matplotlib.patches import Circle # We'll be drawing collections of circular patch objects first
from matplotlib.collections import PatchCollection
```

```python
fig, ax = plt.subplots() # Creating a figure and an axis for a subplot, by default I'm getting only 1 sub-plot here

# Now creating a collection of the patch objects
patches = [Circle((0.1,0.1),0.1), Circle((0.3,0.3),0.1), Circle((0.5,0.5),0.1)]

# Creating a patch collection
collection = PatchCollection(patches, edgecolor='blue', facecolor='yellow', alpha=0.5)

ax.add_collection(collection)

ax.set_xlim(0,1)
ax.set_ylim(0,1)
ax.set_aspect('equal')

plt.show()
```

```python
# That's just 3 circles - Huh? That's it?
# No, we can use more!
```

```python
fig, ax = plt.subplots()
patches = []
for i in range(1, 101):
  patches.append(Circle((i*0.1, i*0.1),0.1))

collection = PatchCollection(patches, edgecolor='black', facecolor='blue', alpha = 1.0)

ax.add_collection(collection)
ax.set_xlim(0,15)
ax.set_ylim(0,15)
ax.set_aspect('equal')
plt.show()
```

```python
# That is a line. What?
# Well look closely. It is 100 circles arranged together and so they resemble a line.
# You can create a line, sure. Well, now we created a line with circles.
```

```python
# What if you want different properties for each patch?
```

```python
from matplotlib.patches import Rectangle # Making patches of rectangles
import numpy as np
```

```python
fig, ax = plt.subplots()

patches = [Rectangle((0.1*i, 0.1*i), 0.1, 0.1) for i in range(5)]
collection = PatchCollection(patches, edgecolor='black')

colors = np.linspace(0,1,len(patches))

collection.set_array(colors)

ax.add_collection(collection)
ax.set_xlim(0,1)
ax.set_ylim(0,1)
ax.set_aspect('equal')


plt.show()
```
